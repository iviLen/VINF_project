### Document and collection frequency - whole section names

```python
def document_freq_collection_freq_section(text):
    st = time.time()
    title = ""
    df_dict = {}
    cf_dict = {}
    
    for line in text:
        tmp = line.split("\t")
        title_act = tmp[-1]
        section = tmp[0].lower()
        
        if title == "":
            title = title_act
            tmp_dictionary = {}
            tmp_dictionary[section] = 1  
        elif title != title_act:
            
            for key, item in tmp_dictionary.items():
                
                if df_dict.get(key):
                    df_dict[key] = df_dict[key] + 1
                else:
                    df_dict[key] = 1
                    
                if cf_dict.get(key):
                    cf_dict[key] = cf_dict[key] + item
                else:
                    cf_dict[key] = item
                    
            title = title_act
            tmp_dictionary = {}
            tmp_dictionary[section] = 1   
        else:
            if tmp_dictionary.get(section):
                tmp_dictionary[section] = tmp_dictionary[section] + 1
            else:
                tmp_dictionary[section] = 1
        
    for key, item in tmp_dictionary.items():
        
        if df_dict.get(key):
            df_dict[key] = df_dict[key] + 1
        else:
            df_dict[key] = 1

        if cf_dict.get(key):
            cf_dict[key] = cf_dict[key] + item
        else:
            cf_dict[key] = item   
    
      
    end = time.time()
    print ("total time", end-st)
    return df_dict, cf_dict
```

### Document and collection frequency - token

```python
def document_freq_and_collection_freq_tokenize(text):
    st = time.time()
    title = ""
    count_words = 0
    df_dict = {}
    cf_dict = {}
    wnl = nltk.WordNetLemmatizer()
    
    for line in text:
        tmp = line.split("\t")
        title_act = tmp[-1]
        section = tmp[0].lower()
        tokens = nltk.word_tokenize(section)
        
        if title == "":
            title = title_act
            tmp_dictionary = {}
            for token in tokens:
                word = wnl.lemmatize(token)
                tmp_dictionary[word] = 1 
                count_words = count_words + 1
        elif title != title_act:
            
            for key, item in tmp_dictionary.items():
                
                if df_dict.get(key):
                    df_dict[key] = df_dict[key] + 1
                else:
                    df_dict[key] = 1
                    
                if cf_dict.get(key):
                    cf_dict[key] = cf_dict[key] + item
                else:
                    cf_dict[key] = item
                    
            title = title_act
            tmp_dictionary = {}
            for token in tokens:
                word = wnl.lemmatize(token)
                tmp_dictionary[word] = 1  
                count_words = count_words + 1
        else:
            for token in tokens:
                word = wnl.lemmatize(token)
                count_words = count_words + 1
                if tmp_dictionary.get(word):
                    tmp_dictionary[word] = tmp_dictionary[word] + 1
                else:
                    tmp_dictionary[word] = 1
        
    for key, item in tmp_dictionary.items():
        
        if df_dict.get(key):
            df_dict[key] = df_dict[key] + 1
        else:
            df_dict[key] = 1

        if cf_dict.get(key):
            cf_dict[key] = cf_dict[key] + item
        else:
            cf_dict[key] = item   
    
      
    end = time.time()
    print ("total time", end-st)
    return df_dict, cf_dict, count_words
```

### Statistics

```python
def stats(text):
    
    number_of_pages = 0
    number_of_sections = 0
    title = ""
    maximum_title = ""
    maximum_section = 0
    tmp_number_section = 0
    dict_sum = {}
    dict_title = {}
    
    for line in text:
        tmp = line.split("\t")
        title_act = tmp[-1]
        number_of_sections = number_of_sections + 1
        
        if title == "":
            tmp_number_section = 1
            title = title_act
            number_of_pages = number_of_pages + 1
            
        elif title != title_act:
            
            if maximum_section < tmp_number_section:
                maximum_section = tmp_number_section
                maximum_title = title
                
            if dict_sum.get(tmp_number_section):
                dict_sum[tmp_number_section] = dict_sum[tmp_number_section] + 1
            else:
                dict_sum[tmp_number_section] = 1
                
            if dict_title.get(tmp_number_section):
                k = len(title)
                title = title[:k-1]
                dict_title[tmp_number_section].append(title)
            else:
                k = len(title)
                title = title[:k-1]
                l = []
                l.append(title)
                dict_title[tmp_number_section] = l
            
            tmp_number_section = 1
            title = title_act
            number_of_pages = number_of_pages + 1
        else:
            tmp_number_section = tmp_number_section + 1
    
    if maximum_section < tmp_number_section:
        maximum_section = tmp_number_section
        maximum_title = title
    
    if dict_sum.get(tmp_number_section):
        dict_sum[tmp_number_section] = dict_sum[tmp_number_section] + 1
    else:
        dict_sum[tmp_number_section] = 1
    
    if dict_title.get(tmp_number_section):
        k = len(title)
        title = title[:k-1]
        dict_title[tmp_number_section].append(title)
    else:
        k = len(title)
        title = title[:k-1]
        l = []
        l.append(title)
        dict_title[tmp_number_section] = l
    
    k = len(maximum_title)
    maximum_title = maximum_title[:k-1]
    most_frequent_number = max(dict_sum, key=dict_sum.get)
    
    print("Number of records (sections in dataset): " + str(number_of_sections))
    print("Number of pages: " + str(number_of_pages))
    print("Average number of sections per page: " + str(number_of_sections/number_of_pages))
    print("Maximum number of section per page: " + str(maximum_section) + " (Title: " + maximum_title + ")")
    print("Most frequent number of sections: " + str(most_frequent_number) + " in " + str(dict_sum[most_frequent_number]) + " pages")
    
    return dict_sum, dict_title
    
```

### Document and collection frequency - indexing

```python
import ast
```

```python
from elasticsearch import Elasticsearch
from elasticsearch import helpers
```

```python
es = Elasticsearch()
```

```python
def index_frequencies(dictionary, name_of_index, section_or_token):    
    st = time.time()
    tmp_list = []
    count = 1
    for key, item in dictionary.items():
        actions = {
                "_index": name_of_index,
                "_type": "doc",
                "_id": count,
                "_source": {
                    section_or_token: key,
                    "number": item,
                    }
              }
        tmp_list.append(actions)
        if (count % 100000 == 0):
            helpers.bulk(es, tmp_list)
            tmp_list.clear()
        if (count % 1000000 == 0):
            print(count)
        count = count + 1   

    if ((count-1) % 100000 != 0):
        helpers.bulk(es, tmp_list)

    end = time.time()
    print ("total time", end-st)
```

```python
f_df = open('data/document_freq.txt', encoding="utf8") 
text_df = f_df.read()
f_df.close()
```

```python
df_dict = ast.literal_eval(text_df)
```

```python
index_frequencies(df_dict, "df_sections", "section")
```
