```python
import re
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch()
```

```python
def search_section(word):
    body = {
        "from":0,
        "size":100,
        "track_total_hits": True,
        "query": {
            "match": {
                "section":word
            }
        }
    }

    result = es.search(index="all_data_final", body=body)
    flag = 0
    count = 0
    
    if result['hits']['total']['value'] > 15:
        print("Too many results (15+). Specify your search term please.")
        flag = 1
    else:
        print("Number of results: " + str(result['hits']['total']['value']))
    
    for match in result['hits']['hits']:
        if flag == 1:
            count = count + 1
        if count > 15:
            break
        print("---------------------------------------")
        print("*Title of Wikipedia page:*")
        print(match['_source']['title'])
        print("*Section:*")
        print(match['_source']['section']+"\n")
        print("*Hierarchy:*")
        print(match['_source']['hierarchy'])
        
    
```

```python
def search_title(word):
    body = {
        "from":0,
        "size":10000,
        "track_total_hits": True,
        "query": {
            "match": {
                "title":word
            }
        }
    }

    result = es.search(index="all_data_final", body=body)
    flag = 0
    count = 0
    
    if result['hits']['total']['value'] > 15:
        print("Too many results (15+). Specify your search term please.")
        flag = 1
    else:
        print("Number of results: " + str(result['hits']['total']['value']))
    
    for match in result['hits']['hits']:
        #if flag == 1:
            #count = count + 1
        if count > 15:
            break
        print("---------------------------------------")
        print("*Title of Wikipedia page:*")
        print(match['_source']['title'])
        print("*Section:*")
        print(match['_source']['section']+"\n")
        print("*Hierarchy:*")
        print(match['_source']['hierarchy'])
       
```

### Search by section
#### Write the search term in quotation marks

```python
word = "bratislava"
search_section(word)
```

### Search by title
#### Write the search term in quotation marks

```python
word = "Colin edwin"
search_title(word)
```

### Document frequency and collection frequency - whole section name

```python
def document_frequency_section(word):
    body = {
        "from":0,
        "size":100,
        "track_total_hits": True,
        "sort" : [
            { "number" : "desc" },
            "_score"
        ],
        "query": {
            "match": {
                "section":word
            }
        }
    }

    result = es.search(index="df_sections", body=body)
    flag = 0
    count = 0
    
    if result['hits']['total']['value'] > 5:
        print("Too many results (5+). Specify your search term please.")
        flag = 1
    else:
        print("Number of results: " + str(result['hits']['total']['value']))
    
    for match in result['hits']['hits']:
        if flag == 1:
            count = count + 1
        if count > 5:
            break
        print("---------------------------------------")
        print("*Section:*")
        print(match['_source']['section'])
        print("*Number of pages where the searched section is present:*")
        print(match['_source']['number'])
        
```

```python
def collection_frequency_section(word):
    body = {
        "from":0,
        "size":100,
        "track_total_hits": True,
        "sort" : [
            { "number" : "desc" },
            "_score"
        ],
        "query": {
            "match": {
                "section":word
            }
        }
    }

    result = es.search(index="cf_sections", body=body)
    flag = 0
    count = 0
    
    if result['hits']['total']['value'] > 5:
        print("Too many results (5+). Specify your search term please.")
        flag = 1
    else:
        print("Number of results: " + str(result['hits']['total']['value']))
    
    for match in result['hits']['hits']:
        if flag == 1:
            count = count + 1
        if count > 5:
            break
        print("---------------------------------------")
        print("*Section:*")
        print(match['_source']['section'])
        print("*Number of occurance of the searched section in collection:*")
        print(match['_source']['number'])
```

### Document frequency - whole section name
#### Write the search term in quotation marks

```python
word = "see also"
document_frequency_section(word)
```

### Collection frequency - whole section name
#### Write the search term in quotation marks

```python
word = "see also"
collection_frequency_section(word)
```
