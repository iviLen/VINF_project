```python
import re
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch()
```

```python
def search_section(word):
    body = {
        "from":0,
        "size":10000,
        "track_total_hits": True,
        "query": {
            "match": {
                "section":word
            }
        }
    }

    result = es.search(index="whole_data_final", body=body)
    
    if result['hits']['total']['value'] > 10000:
        print("Number of results: 10 000+")
    else:
        print("Number of results: " + str(result['hits']['total']['value']))
    
    for match in result['hits']['hits']:
        print("---------------------------------------")
        print("*Title of Wikipedia page:*")
        print(match['_source']['title'])
        print("*Section:*")
        print(match['_source']['section']+"\n")
        print("*Hierarchy:*")
        print(match['_source']['hierarchy'])
        
    
```

### Searched word - write in quotation marks

```python
word = "References"
search_section(word)
```
