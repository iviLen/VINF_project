```python
import re
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch()
```

```python
def searching(word):
    body = {
        "from":0,
        "size":10000,
        "track_total_hits": True,
        "query": {
            "match": {
                "section":word
            }
        }
    }

    result = es.search(index="section_headers", body=body)
    
    for match in result['hits']['hits']:
        print(match['_source']['section'])
    
```

### Searched word - write in quotation marks

```python
word = "history"
searching(word)
```
