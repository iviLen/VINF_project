```python
import re
import pandas as pd
import bz2
import time
from datetime import datetime
import nltk
```

```python
def document_freq_collection_freq_section(text):
    st = time.time()
    title = ""
    df_dict = {}
    cf_dict = {}
    
    for line in text:
        tmp = line.split("\t")
        title_act = tmp[-1]
        section = tmp[0].lower()
        
        if title == "":
            title = title_act
            tmp_dictionary = {}
            tmp_dictionary[section] = 1  
        elif title != title_act:
            
            for key, item in tmp_dictionary.items():
                
                if df_dict.get(key):
                    df_dict[key] = df_dict[key] + 1
                else:
                    df_dict[key] = 1
                    
                if cf_dict.get(key):
                    cf_dict[key] = cf_dict[key] + item
                else:
                    cf_dict[key] = item
                    
            title = title_act
            tmp_dictionary = {}
            tmp_dictionary[section] = 1   
        else:
            if tmp_dictionary.get(section):
                tmp_dictionary[section] = tmp_dictionary[section] + 1
            else:
                tmp_dictionary[section] = 1
        
    for key, item in tmp_dictionary.items():
        
        if df_dict.get(key):
            df_dict[key] = df_dict[key] + 1
        else:
            df_dict[key] = 1

        if cf_dict.get(key):
            cf_dict[key] = cf_dict[key] + item
        else:
            cf_dict[key] = item   
    
      
    end = time.time()
    print ("total time", end-st)
    return df_dict, cf_dict
```

```python
def document_freq_and_collection_freq_tokenize(text):
    st = time.time()
    title = ""
    df_dict = {}
    cf_dict = {}
    wnl = nltk.WordNetLemmatizer()
    
    for line in text:
        tmp = line.split("\t")
        title_act = tmp[-1]
        section = tmp[0].lower()
        tokens = nltk.word_tokenize(section)
        
        if title == "":
            title = title_act
            tmp_dictionary = {}
            for token in tokens:
                word = wnl.lemmatize(token)
                tmp_dictionary[word] = 1  
        elif title != title_act:
            
            for key, item in tmp_dictionary.items():
                
                if df_dict.get(key):
                    df_dict[key] = df_dict[key] + 1
                else:
                    df_dict[key] = 1
                    
                if cf_dict.get(key):
                    cf_dict[key] = cf_dict[key] + item
                else:
                    cf_dict[key] = item
                    
            title = title_act
            tmp_dictionary = {}
            for token in tokens:
                word = wnl.lemmatize(token)
                tmp_dictionary[word] = 1    
        else:
            for token in tokens:
                word = wnl.lemmatize(token)
                if tmp_dictionary.get(word):
                    tmp_dictionary[word] = tmp_dictionary[word] + 1
                else:
                    tmp_dictionary[word] = 1
        
    for key, item in tmp_dictionary.items():
        
        if df_dict.get(key):
            df_dict[key] = df_dict[key] + 1
        else:
            df_dict[key] = 1

        if cf_dict.get(key):
            cf_dict[key] = cf_dict[key] + item
        else:
            cf_dict[key] = item   
    
      
    end = time.time()
    print ("total time", end-st)
    return df_dict, cf_dict
```

```python
def average_number_of_sections_per_page(text):
    
    number_of_pages = 0
    number_of_sections = 0
    title = ""
    
    for line in text:
        tmp = line.split("\t")
        title_act = tmp[-1]
        number_of_sections = number_of_sections + 1
        if title != title_act:
            title = title_act
            number_of_pages = number_of_pages + 1
    
    print("Number of records (sections in dataset): " + str(number_of_sections))
    print("Number of pages: " + str(number_of_pages))
    print("Average number of sections per page: " + str(number_of_sections/number_of_pages))
    
```
